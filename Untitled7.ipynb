```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow.keras import layers,models
```

# DATA PREPROCESSING



It was the structure of data_set

```python
train_dir = "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train"

valid_dir = "/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid"

test_dir = "/kaggle/input/new-plant-diseases-dataset/test/test"
```

```python
print("Number of classes:", len(os.listdir(train_dir)))
print(os.listdir(train_dir))
```

Now we will make sure that image size is 224 by 224

```python
IMG_SIZE = 224
BATCH_SIZE = 126
```

```python
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(train_dir, batch_size=BATCH_SIZE,image_size=(IMG_SIZE, IMG_SIZE))
valid_dataset= tf.keras.preprocessing.image_dataset_from_directory(valid_dir, batch_size=BATCH_SIZE,image_size=(IMG_SIZE, IMG_SIZE))
```

```python
#NOW WE WILL NORMALIZE THE IMAZES BETWEEN ZERO TO ONE SCALE SO COMPUTING BECOMES EASY
normalization = tf.keras.layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization(x), y))
valid_dataset = valid_dataset.map(lambda x, y: (normalization(x), y))
print("Images are normalized in scale [0,1]")
```

```python
# WE WILL MAKE DATA A BIT MORE RADOMIZED SO THAT THE MODEL DOSENT GET OVERFIT
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2)
])
```

```python
# A COMMON PRACTICE TO INCREASE SYSTEM RESOURCES EFFICIENCY
AUTOTUNE = tf.data.AUTOTUNE

train_dataset = train_dataset.prefetch(AUTOTUNE)
valid_dataset = valid_dataset.prefetch(AUTOTUNE)
```

**MODELS**

```python
# here i will define my model arcitecture
my_model=models.Sequential([
    tf.keras.Input(shape=(224,224,3)),
    data_augmentation,
    normalization,
    layers.Conv2D(32, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),


    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),


    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),


    layers.Conv2D(256, (3,3), activation='relu', padding='same'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),


    tf.keras.layers.GlobalAveragePooling2D(),

    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),

    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),

    layers.Dense(38, activation='softmax')
])
my_model.summary()
```

```python
# here we have pre_trained EFFICIENTNETB0
# WE  will do transform learning instead whole traning
pre_trained_effnet = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(224, 224, 3)
)

pre_trained_effnet.trainable = False
model=keras.Sequential([
    data_augmentation,
    normalization,
    pre_trained_effnet,
    layers.GlobalAveragePooling2D(),
    layers.Dense(38, activation='softmax')
])
model.summary()
```

NOW COMES THE TRAINING

```python
# we will first compile and train my model
my_model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy,
    metrics=['accuracy']
)
history_my_model=my_model.fit(
    train_dataset,
    validation_data=valid_dataset,
    batch_size=BATCH_SIZE,
    epochs=10,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        )
    ]
)
```
